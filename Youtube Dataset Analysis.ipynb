```python
df.dropna(axis=0,inplace=True)
```

```python
df.head(20)
```

```python
sns.heatmap(df.isnull())
```

```python
df.tail(5)
```

```python
df['Rank']=df['Rank'].str[0:-2]
```

```python
df.tail(5)
```

```python
df['Rank']=df['Rank'].str.replace(',','').astype('int')
```

```python
df.tail(5)
```

```python
df.dtypes
```

```python
df['Video Uploads']=df['Video Uploads'].astype('int')
```

```python
df['Subscribers']=df['Subscribers'].astype('int')
```

```python
df['Grade'].unique()
```

```python
df['Grade']=df['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1})
```

```python
df.head(10)
```

```python
df['Avg Views']=df['Video views']/df['Video Uploads']
```

```python
df.head()
```

```python
s= df.groupby(['Channel name'], as_index=False)['Avg Views'].sum().sort_values(by='Avg Views', ascending=False).head(5)

sns.set(rc={'figure.figsize':(12,5)})
sns.barplot(data = s, x = 'Channel name',y= 'Avg Views')
```

cocoz toon is having the maximum audience retention capability and NBA LeB is performing the worst.
